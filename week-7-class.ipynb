```python
%matplotlib inline
import networkx as nx
import random
from collections import defaultdict
import math
import matplotlib.pyplot as plt
infinity = float('inf')
```

```python
def visualise(G, W, layout=nx.spring_layout):
    G_out = nx.DiGraph()
    for u in G.keys():
        G_out.add_node(str(u))
        for v in G[u]:
            G_out.add_edge(str(u), str(v), weight=str(W[u, v]))
    pos = layout(G_out)
    nx.draw_networkx(G_out, pos=pos)
    nx.draw_networkx_edge_labels(G_out, pos=pos, edge_labels=W)

def graph_from_edges(edges):
    # assume edges = list of triples (u, v, w)
    nodes = sorted(list(set([u for u, v, w in edges]+[v for u, v, w in edges])))
    G = defaultdict(list)
    W = {}
    for u, v, w in edges:
        G[u].append(v)
        W[u, v] = w
    return G, W
```

```python
def dijkstra(G, W, start):
    predecessor = {start: None}   # predecessor subgraph (start has no predecessor)
    q = {start: 0}                # priority queue for next vertices, distance to start is 0
    shortest = {}                 # shortest paths found (none at the start)
    # Repeatedly evaluate this cell until it says Finished!
    while q: # are there any vertices to add?
        # Find the lowest priority (note: this is where the data structure is inefficient!)
        d, vertex = min((d, vertex) for vertex, d in q.items())
        del q[vertex]
        # Add this to the set of vertices we've finished processing
        shortest[vertex] = d
        # Now update the list of candidate vertices by iterating through its neighbours
        for next_vertex in G[vertex]:
            if next_vertex in shortest: # don't add ones we've already finished with
                continue
            c = q.get(next_vertex, infinity) # get the current best found if there is one, or infinity if not
            e = d+W[vertex, next_vertex] # distance to next_vertex if you take this route
            if e<c: # if it's better, replace it in the queue and predecessor subgraph
                q[next_vertex] = e
                predecessor[next_vertex] = vertex
    return shortest, predecessor

def shortest_path_subgraph(G, W, target, predecessor):
    SG = {}
    SW = {}
    while predecessor[target] is not None:
        source = predecessor[target]
        SG[source] = [target]
        SW[source, target] = W[source, target]
        target = source
    return SG, SW
```

### Q1: Give an example to show that Dijkstra's algorithm fails on graphs if you allow negative weights. Why does it fail?

Consider the graph:
- A → B (weight 1)
- A → C (weight 0)
- B → C (weight -2)
- C → D (weight 3)

Dijkstra from A: picks C first (distance 0) and marks it as done. Later discovers A→B→C has distance -1, but C is already finalised — so it never updates C.

**Dijkstra result (wrong):** A=0, B=1, C=0, D=3

**Correct result:** A=0, B=1, C=-1, D=2

**Why it fails:** Dijkstra relies on the greedy assumption that once a vertex is processed, its shortest distance is final. With negative weights, a later-discovered path through a negative edge can be shorter than the already-finalised distance, violating this assumption.

```python
# Verify Q1 example
edges_q1 = [
    ('A', 'B', 1),
    ('A', 'C', 0),
    ('B', 'C', -2),
    ('C', 'D', 3),
]

print("Dijkstra (wrong):")
s_d, _ = dijkstra(*graph_from_edges(edges_q1), 'A')
print(s_d)

print("\nBellman-Ford (correct):")
s_bf, p_bf = path_gen(edges_q1, 'A')
print(s_bf)

visualise(*graph_from_edges(edges_q1))
```

### Q3: Bellman-Ford algorithm

The algorithm (implemented as `path_gen` above) works as follows:

1. Initialise `shortest[start] = 0`, all other vertices unknown.
2. Repeat |V| - 1 times: **relax every edge** in the graph. Relaxing edge (u, v, w) means: if `shortest[u] + w < shortest[v]`, update `shortest[v]` and `predecessor[v]`.
3. Do **one more pass** over all edges. If any edge can still be relaxed, a **negative cycle** exists — report an error.

**Why |V| - 1 iterations?** A shortest path in a graph with |V| vertices has at most |V| - 1 edges. Each iteration of relaxation guarantees that paths using one more edge are found. So after |V| - 1 iterations, all shortest paths (if they exist) are found.

**How to detect a problem?** If the |V|-th pass still finds an improvement, it means a negative cycle exists — you can keep reducing the distance forever, so no finite shortest path exists.

```python
# Verify Q3: Bellman-Ford works on Q1 example (negative weights, no negative cycle)
print("Bellman-Ford on edges_q1:")
s, p = path_gen(edges_q1, 'A')
print("Shortest distances:", s)
print("Path A to D:", find_path(p, 'A', 'D'))

print("\nBellman-Ford on edges_problematic (has negative cycle):")
s, p = path_gen(edges_problematic, 'A')
print("Shortest distances:", s)
```

### Q4: What is the complexity of this algorithm?

- **Outer loop:** |V| iterations (|V|-1 relaxation + 1 detection)
- **Inner loop:** iterates over all |E| edges
- **Each relaxation:** O(1)

**Total time complexity: O(|V| · |E|)**

For comparison, Dijkstra's algorithm with a binary heap is O((|V| + |E|) log |V|), which is faster but cannot handle negative weights.
